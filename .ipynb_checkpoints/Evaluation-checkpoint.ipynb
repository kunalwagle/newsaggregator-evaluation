Queries for getting articles

```python
guardianQuery = clusters.aggregate([
    {
        "$unwind":"$Articles"
    },
    {
        "$lookup":{
            "from":"articles",
            "localField":"Articles",
            "foreignField":"_id",
            "as":"article"
        }
    },
    {
        "$match":{
            "article.Source":"the-guardian-uk"
        }
    },
    {"$group": { "_id": "$_id"}},
    {"$project": { "_id": 1,"Articles": "$_id"} }
])

guardianSummaries = list(map(lambda art: art["Articles"], list(guardianQuery)))
```

```python
dmQuery = clusters.aggregate([
    {
        "$unwind":"$Articles"
    },
    {
        "$lookup":{
            "from":"articles",
            "localField":"Articles",
            "foreignField":"_id",
            "as":"article"
        }
    },
    {
        "$match":{
            "article.Source":"daily-mail"
        }
    },
    {"$group": { "_id": "$_id"}},
    {"$project": { "_id": 1,"Articles": "$_id"} }
])

dmSummaries = list(map(lambda art: art["Articles"], list(dmQuery)))
```

```python
dmgSummaries = list(set(dmSummaries) & set(guardianSummaries))

dmgQuery = clusters.find({
    "$and":[
        {
            "_id":{
                "$in":dmgSummaries
            }
        }, 
        {
            "Articles.2":{
                "$exists":False
            }
        }
    ]
})

dmgArticles = list(dmgQuery)
```

```python
import json

dmgsumms = list(map(lambda art: json.loads(art["Summaries"]), dmgArticles))

dmgfilteredsumms = []

for val in dmgsumms:
    for key in val:
        if ("dailymail" in key) & ("guardian" in key):
            dmgfilteredsumms.append(val[key])
```

```python
dma = []
ga = []
for art in dmgArticles:
    twoarticles = articles.find({"_id":{"$in":art['Articles']}})
    for val in twoarticles:
        if val['Source']=='daily-mail':
            dma.append(val)
        else:
            ga.append(val)
```

```python
counter = 1
totaldm = 0;
totalg = 0;
total = 0;

for summary in dmgfilteredsumms:
    sources = list(map(lambda sentence: sentence["source"], summary))
    articles = dmgArticles[counter-1]
    tdm = sources.count('daily-mail')
    tg = sources.count('the-guardian-uk')
    t = len(sources)
    dmcount = (sources.count('daily-mail')/len(sources))*100
    guardiancount = (sources.count('the-guardian-uk')/len(sources))*100
    dmlen = len(dma[counter-1]['Body'])
    glen = len(ga[counter-1]['Body'])
    print("Article", counter, "DM:", dmcount, "G:", guardiancount, "Total:",len(sources), "DM:", dmlen, "G:", glen)
    counter = counter + 1
    
print("DM Total:", (totaldm/total)*100, "G Total:", (totalg/total)*100)
```

```python
def getSummaries( source ):
    query = clusters.aggregate([
        {
            "$unwind":"$Articles"
        },
        {
            "$lookup":{
                "from":"articles",
                "localField":"Articles",
                "foreignField":"_id",
                "as":"article"
            }
        },
        {
            "$match":{
                "article.Source":source
            }
        },
        {"$group": { "_id": "$_id"}},
        {"$project": { "_id": 1,"Articles": "$_id"} }
    ])

    return list(map(lambda art: art["Articles"], list(query)))
```

```python
def getIntersection(first, second):
    summaries = list(set(first) & set(second))

    query = clusters.find({
        "$and":[
            {
                "_id":{
                    "$in":summaries
                }
            }, 
            {
                "Articles.2":{
                    "$exists":False
                }
            }
        ]
    })

    return list(query)
```

```python
def extractSummaries(articles):
    summs = list(map(lambda art: json.loads(art["Summaries"]), articles))

    filteredsumms = []

    for val in summs:
        for key in val:
            if ("dailymail" in key) & ("guardian" in key):
                filteredsumms.append(val[key])
                
    return filteredsumms
```
